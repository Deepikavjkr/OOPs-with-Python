<h3>Slicing</h3>
In this notebook we take a look at a very useful feature in Python called __slicing__.

Slicing enables us to extract subsets of a sequence.  In this notebook we study slicing as it applies to 
extracting portions of a string. 

The general form is `str_variable[start:end:step]`   
&nbsp;&nbsp;&nbsp;&nbsp;`str_variable` :  is any string variable or string literal  
&nbsp;&nbsp;&nbsp;&nbsp;`start`: is an integer representing the starting index  
&nbsp;&nbsp;&nbsp;&nbsp;`end`: is an integer representing the end index.  The last valid integer before the end index will be the last value returned  
&nbsp;&nbsp;&nbsp;&nbsp;`step`: is an integer representing the step value (the increment)


```python
my_str = 'abcdefghij'
print('my_str', my_str)
```

The figure below provides the valid indexes (one negative and one positive) for each one of the characters in `my_str`



Evaluate each of the slicing statements below.

```python
print('my_str[0:3]=', my_str[0:3]) 
```

```python
print('my_str[4:8]=', my_str[4:8]) 
```

```python
print('my_str[2:5]=', my_str[2:5]) 
```

```python
print('my_str[-5:-3]=', my_str[-5:-3]) 
```

```python
print('my_str[:3]=', my_str[:3]) 
```

```python
print('my_str[4:]=', my_str[4:]) 
```

```python
print('my_str[:]=', my_str[:])
```

The figure below provides the valid indexes (one negative and one positive) for each one of the characters in `my_str`



Evaluate each of the slicing statements below.

If the character at the start index comes after (or is) the character at the end index and the step is positive, nothing is printed

```python
print('my_str[-8:2]=', my_str[-8:2])
```

If the start index is less than `-n` and `step` is positive, slicing starts at index 0

```python
print('my_str[-18:3]=', my_str[-18:3])
```

If the start index is less than `-n` and `step` is negative, nothing is printed

```python
print('my_str[-18:3:-1]=', my_str[-18:3:-1])
```

If the end value is greater than the length of the string, all the characters to the end are displayed

```python
print('my_str[4:100]=', my_str[4:100]) 
```

```python
print('my_str[2:-2:2]=', my_str[2:-2:2]) 
```

```python
print('my_str[::2]=', my_str[::2])
```

You can print a string in reverse by omitting the `start` and `end` index values and using a `step` value of `-1`

```python
print('my_str[::-1]=', my_str[::-1]) 
```
